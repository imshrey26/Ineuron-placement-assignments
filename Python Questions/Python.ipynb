Q1: Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word?

```python
def highest_frequency_word_length(string):
    """
    Takes a string as input and return the most frequently occuring word in inputted string
    and its's length
    """
    
    words = string.split()

    word_freq = {}
    for word in words:
        word_freq[word] = word_freq.get(word, 0) + 1

    max_freq = max(word_freq.values())

    max_freq_word_length = max(len(word) for word, freq in word_freq.items() if freq == max_freq)

    return max_freq_word_length

```

Question 2: -
Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

```python
from collections import Counter

def is_valid_string(string):
    # Count the frequency of each character
    char_freq = Counter(string)

    # Count the frequency of frequencies
    freq_freq = Counter(char_freq.values())

    # If there is only one frequency, the string is valid
    if len(freq_freq) == 1:
        return "YES"

    # If there are more than two frequencies, the string is not valid
    if len(freq_freq) > 2:
        return "NO"

    # If there are exactly two frequencies, check if we can remove one character to make it valid
    freq1, count1 = freq_freq.most_common(1)[0]
    freq2, count2 = freq_freq.most_common(2)[1]

    # If the difference between the two frequencies is 1 and the frequency with the higher count occurs only once,
    # or if the frequency with the lower count is 1, the string is valid
    if (freq1 == 1 and count1 == 1) or (freq2 - freq1 == 1 and count2 == 1):
        return "YES"

    return "NO"


# Example test cases
test_cases = [
    "abcde",   # Valid string: All characters occur once
    "aabbbcc", # Valid string: All characters occur the same number of times (2)
    "aabbccd", # Valid string: Removing 'd' will make all characters occur the same number of times (2)
    "aabbccdd",# Not valid: More than one character needs to be removed to make it valid
]

for test_case in test_cases:
    result = is_valid_string(test_case)
    print(f"{test_case} -> {result}")
```

Q4: Write a program to download the data from the link given below and then read the data and convert the into
the proper structure and return it as a CSV file.

Link - https://data.nasa.gov/resource/y77d-th95.json

```python
import pandas as pd
from pandas import json_normalize

url_2 = 'https://data.nasa.gov/resource/y77d-th95.json'

def json_to_csv(url,filename):
    """ Takes url containing json data and converts it to dataframe and then exports
        the dataframe to csv file, filename should be provided as second input """
    
    responses = requests.get(url)
    data = responses.json()

    df = pd.DataFrame(data)
    print('Old Dataframe')
    print(df)
    
    for col in df.columns:
        if type(df[col][0]) == dict:
            df_new = pd.concat([df.drop(col, axis=1),
                              json_normalize(df[col]).add_prefix(f'{col}.')],
                             axis=1)
    print('New DataFrame')
    print(df_new)
    f = filename
    csv_file = df_new.to_csv(f, index=False, encoding='utf-8')
    print('Succesfully imported the excel file')
    return csv_file
    
    
```

Q5: Get all Pokemons whose spawn time is less than 300 seconds.

```python
def convert_to_seconds(time_str):
    if pd.isnull(time_str):
        return np.nan
    
    else:
        time = time_str.split(':')
        minutes,seconds = time[0],time[1]
        total_seconds = int(minutes) * 60 + int(seconds)
        return total_seconds
```

```python
pokemon_data['spawn_seconds'] = pokemon_data['spawn_time'].apply(convert_to_seconds)
```

Q10: Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.

```python
import re

def count_pos(phrase):
    # Convert the phrase to lowercase for easier matching
    phrase = phrase.lower()

    # Define patterns for different parts of speech
    noun_pattern = r'\b(?:[a-z]+)\b'  # Matches any word as a noun
    verb_pattern = r'\b(?:[a-z]+ed|[a-z]+ing|[a-z]+s)\b'  # Matches common verb forms
    pronoun_pattern = r'\b(?:i|you|he|she|it|we|they|me|him|her|us|them)\b'  # Matches common pronouns
    adjective_pattern = r'\b(?:[a-z]+y|[a-z]+ish|[a-z]+ful)\b'  # Matches common adjective forms

    # Initialize counts for each part of speech
    counts = {'Noun': 0, 'Verb': 0, 'Pronoun': 0, 'Adjective': 0}

    # Count the occurrences of each part of speech
    counts['Noun'] = len(re.findall(noun_pattern, phrase))
    counts['Verb'] = len(re.findall(verb_pattern, phrase))
    counts['Pronoun'] = len(re.findall(pronoun_pattern, phrase))
    counts['Adjective'] = len(re.findall(adjective_pattern, phrase))

    return counts

# Example phrase
phrase = "Data Science is the best and the hardest jobs in India"

# Count the parts of speech in the phrase
pos_counts = count_pos(phrase)

# Print the counts
print(pos_counts)
```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def count_pos(phrase):
    # Tokenize the phrase into words
    words = word_tokenize(phrase)

    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)

    # Initialize counts for each part of speech
    counts = {'Noun': 0, 'Verb': 0, 'Pronoun': 0, 'Adjective': 0}

    # Count the occurrences of each part of speech
    for word, pos in tagged_words:
        if pos.startswith('NN'):  # Noun
            counts['Noun'] += 1
        elif pos.startswith('VB'):  # Verb
            counts['Verb'] += 1
        elif pos.startswith('PRP'):  # Pronoun
            counts['Pronoun'] += 1
        elif pos.startswith('JJ'):  # Adjective
            counts['Adjective'] += 1

    return counts

```

```python
phrase = "Data Science is the best and the hardest jobs in India"

pos_counts = count_pos(phrase)

print(pos_counts)
```
